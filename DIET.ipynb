```python
import pandas as pd

# Reload the dataset
df = pd.read_csv('X.csv')

# Define the list of nonveg items
nonveg_items = ['egg', 'chicken', 'meat', 'steak', 'fish', 'prawns', 'lamb', 'mutton', 'beef', 'lobster', 'Shrimp', 'crab', 'salmon', 'Tuna', 'Omlette']

# Function to classify food items
def classify_food(item):
    item_lower = str(item).lower()
    if any(nonveg in item_lower for nonveg in nonveg_items):
        return 'Nonveg'
    else:
        return 'Veg'

# Apply classification to create a 'Category' column
df['Category'] = df['RecipeIngredientParts'].apply(classify_food)

# Create separate Veg and Nonveg columns
df['Veg'] = df['RecipeIngredientParts'].where(df['Category'] == 'Veg', '')
df['Nonveg'] = df['RecipeIngredientParts'].where(df['Category'] == 'Nonveg', '')

# Save and inspect the dataset
df.to_csv('Updated_dataset_with_categories.csv', index=False)
print("Veg and Nonveg columns created successfully.")
```

```python
import pandas as pd
from itertools import combinations

# Load datasets (ensure these are Excel files as per your uploaded files)
users_df = pd.read_excel('/content/Calories.xlsx')
meals_df = pd.read_excel('/content/Nonveg_Data.xlsx')

# Default activity multiplier for TDEE
default_activity_factor = 1.2

# Function to calculate BMR
def calculate_bmr(row):
    if row['Gender'] == 'Male':
        return 10 * row['Weight (kg)'] + 6.25 * row['Height (cm)'] - 5 * row['Age'] + 5
    else:  # Female
        return 10 * row['Weight (kg)'] + 6.25 * row['Height (cm)'] - 5 * row['Age'] - 161

# Function to calculate TDEE
def calculate_tdee(row):
    bmr = calculate_bmr(row)
    return bmr * default_activity_factor

# Add calorie intake target and macronutrient targets to users_df
users_df['CalorieIntakeTarget'] = users_df.apply(
    lambda row: calculate_tdee(row) + row['Calories_burnt'], axis=1
)
users_df['ProteinTarget'] = users_df['CalorieIntakeTarget'] * 0.25 / 4  # 1g protein = 4 kcal
users_df['FatTarget'] = users_df['CalorieIntakeTarget'] * 0.30 / 9      # 1g fat = 9 kcal
users_df['CarbsTarget'] = users_df['CalorieIntakeTarget'] * 0.45 / 4    # 1g carbs = 4 kcal
users_df['FiberTarget'] = users_df['CalorieIntakeTarget'] * 0.02 / 2    # Assume 1g fiber = 2 kcal

# Initialize a list to store merged rows
merged_rows = []

# Iterate through users and find meal plans
for _, user in users_df.iterrows():
    calorie_target = user['CalorieIntakeTarget']
    protein_target = user['ProteinTarget']
    fat_target = user['FatTarget']
    carbs_target = user['CarbsTarget']
    fiber_target = user['FiberTarget']

    # Find meal combinations (3-4 meals)
    meals = meals_df[['Name', 'Calories', 'Protein', 'Fat',
                      'Carbohydrate', 'Fiber', 'Category']]
    for num_meals in range(3, 5):  # Plan for 3 or 4 meals
        for meal_combination in combinations(meals.itertuples(index=False), num_meals):
            total_calories = sum(meal.Calories for meal in meal_combination)
            total_protein = sum(meal.Protein for meal in meal_combination)
            total_fat = sum(meal.Fat for meal in meal_combination)
            total_carbs = sum(meal.Carbohydrate for meal in meal_combination)
            total_fiber = sum(meal.Fiber for meal in meal_combination)

            # Check if the combination meets approximate nutritional targets
            if (
                calorie_target * 0.9 <= total_calories <= calorie_target * 1.1 and
                protein_target * 0.9 <= total_protein <= protein_target * 1.1 and
                fat_target * 0.9 <= total_fat <= fat_target * 1.1 and
                carbs_target * 0.9 <= total_carbs <= carbs_target * 1.1 and
                fiber_target * 0.9 <= total_fiber <= fiber_target * 1.1
            ):
                day_meal = [f"{meal.Name} ({meal.Category})" for meal in meal_combination]  # Combine meal names with categories
                merged_rows.append({
                    'User_ID': user['User_ID'],
                    'Gender': user['Gender'],
                    'Age': user['Age'],
                    'Height (cm)': user['Height (cm)'],
                    'Weight (kg)': user['Weight (kg)'],
                    'Duration (min)': user['Duration (min)'],  # Include from Calories.xlsx
                    'Heart_Rate': user['Heart_Rate'],          # Include from Calories.xlsx
                    'Body_Temp': user['Body_Temp'],            # Include from Calories.xlsx
                    'Calories_burnt': user['Calories_burnt'],  # Include from Calories.xlsx
                    'Day_Meal': ", ".join(day_meal),           # Combine meal names with categories into a single string
                    'TotalCalories': total_calories,
                    'TotalProtein': total_protein,
                    'TotalFat': total_fat,
                    'TotalCarbs': total_carbs,
                    'TotalFiber': total_fiber
                })
                break  # Stop after finding the first valid combination

# Create a final DataFrame
merged_df = pd.DataFrame(merged_rows)

# Save the result
merged_df.to_csv('merged_nonveg_meals.csv', index=False)

print("Merged dataset with Non-veg meal plans and categories created successfully.")
```
